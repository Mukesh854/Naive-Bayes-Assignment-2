Q1. A company conducted a survey of its employees and found that 70% of the employees use the
company's health insurance plan, while 40% of the employees who use the plan are smokers. What is the
probability that an employee is a smoker given that he/she uses the health insurance plan?

70% of employees use the health insurance plan (let's call this event A).
Among employees who use the health insurance plan, 40% are smokers (let's call this event B).
We want to find the probability that an employee is a smoker given that they use the health insurance plan, which is 

P(B∣A).

To find this, we can directly use the information provided. Since 40% of employees who use the health insurance plan are smokers,
P(B∣A)=0.40.

So, the probability that an employee is a smoker given that they use the health insurance plan is 40%.

Q2. What is the difference between Bernoulli Naive Bayes and Multinomial Naive Bayes?

. Bernoulli Naive Bayes (BNB) is used for binary features, where each feature can only have two possible values (usually 0 or 1).

. Multinomial Naive Bayes (MNB) is used for features representing counts or frequencies, often with multiple possible outcomes.

. BNB focuses on whether a feature is present or absent, while MNB considers the frequency of each feature.

Bernoulli Naive Bayes doesn't have a built-in mechanism to handle missing values. Common approaches include imputation, ignoring missing values, or treating them as a separate category.

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.metrics import recall_score , accuracy_score , precision_score , f1_score
from sklearn.model_selection import cross_val_score
```

```python
X = df.drop("spam" , axis=1)
y = df["spam"]
```

```python
X.head()
```

```python
bernoulli_nb = BernoulliNB()
multinomial_nb = MultinomialNB()
gaussian_nb = GaussianNB()
```

```python
# here i am making a function 

def calculate_metrics_cv(classifier , x , y):
    accuracy = cross_val_score(classifier , x , y , cv=10 , scoring = "accuracy")
    precision = cross_val_score(classifier ,x , y , cv=10 , scoring = "precision")
    recall = cross_val_score(classifier , x, y , cv=10 , scoring = "recall")
    f1 = cross_val_score(classifier, X, y, cv=10, scoring='f1')
    return accuracy, precision, recall, f1
```

```python
bernoulli_accuracy, bernoulli_precision, bernoulli_recall, bernoulli_f1 = calculate_metrics_cv(bernoulli_nb, X, y)


multinomial_accuracy, multinomial_precision, multinomial_recall, multinomial_f1 = calculate_metrics_cv(multinomial_nb, X, y)


gaussian_accuracy, gaussian_precision, gaussian_recall, gaussian_f1 = calculate_metrics_cv(gaussian_nb, X, y)
```

```python
print("Bernoulli Naive Bayes:")

print("  Accuracy:", bernoulli_accuracy)

print("  Precision:", bernoulli_precision)
print("  Recall:", bernoulli_recall)
print("  F1 score:", bernoulli_f1)

```

Bernoulli Naive Bayes performed the best, likely because it assumes binary features, which align well with the dataset's nature.

Multinomial Naive Bayes followed, as it's suitable for integer feature counts but slightly less so for binary features.

Gaussian Naive Bayes performed the worst since it assumes a Gaussian distribution, which may not fit well with binary features.
